Para começar vou importar uma base de todas as ocorrências com a aviação civil no Brasil.

```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/carlosfab/dsnp2/master/datasets/ocorrencias_aviacao.csv', index_col='codigo_ocorrencia')

df.head()
```

## Resumo

Para ver um resumo das variáveis vou utilizar a função info().

```python
df.info()
```

## Tipo de cada coluna

Podemos usar a função dtypes para encontrar apenas os tipos de cada variável.

```python
df.dtypes
```

## Valores únicos

Vou utilizar a função unique para ver os valores únicos de uma coluna.

```python
df.ocorrencia_tipo.unique()
```

A função acima retorna apenas os valores únicos de uma coluna. Entretanto, há casos em que queremos saber a quantidade de cada valor

```python
df.ocorrencia_tipo.value_counts()
```

E podemos retornar os valores acima em porcentagem dividindo pelos número total dados.

```python
df.ocorrencia_tipo.value_counts() / df.shape[0]
```

## Filtrando valores

Vamos discutir algumas funções para filtrar valores já que existem dezenas de formas diferentes.

### loc

`loc[<linhas>, <colunas>]` 

Vamos supor que queremos saber a latitude e longitude da ocorrência de número `200808073991179`:

```python
df.loc[200808073991179, ['ocorrencia_latitude', 'ocorrencia_longitude']]
```

Vamos agora para um exemplo onde queremos apenas os vôos cujos os tipos de ocorrências são iguais a "FALHA DO MOTOR EM VOO":

```python
df.loc[df['ocorrencia_tipo']=="FALHA DO MOTOR EM VOO"].head()
```

### iloc

`iloc[<numero_linha>, <numero_coluna>]` 

O uso é mais simples pois é necessário apenas os números referentes às linhas e colunas.


```python
df.iloc[:3, :4]
```
